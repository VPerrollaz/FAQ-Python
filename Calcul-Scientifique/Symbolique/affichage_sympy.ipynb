```python
from sympy import *
```

```python
%matplotlib inline
```

```python
init_printing()
```

```python
x, y = symbols("x y")
```

# Comment afficher le graphe d'une fonction sur un intervalle donné

```python
plot(x * x, (x, -1, 1))
```

# Comment régler la taille des affichages

```python
import matplotlib.pyplot as plt

plt.rcParams["figure.figsize"] = (12, 8)
```

```python
plot(x * x, (x, -1, 1))
```

# Comment spécifier la couleur

```python
plot(x * x, (x, -1, 1), line_color="red")
```

# Comment afficher plusieurs fonctions

```python
plot(x * x, x ** 3, (x, -1, 1))
```

# Comment superposer des graphes à priori distincts

```python
p = plot(x * x, (x, -1, 1), line_color="red", show=False)
p.extend(plot(x ** 3, (x, -1, 1), line_color="green", show=False))
p.show()
```

# Comment afficher une courbe paramétrée

```python
plot_parametric(cos(x), sin(x), (x, -pi, pi))
```

# Comment afficher une courbe donnée par une équation

```python
plot_implicit(Eq(10 * x * y, 1), (x, -1, 1), (y, -1, 1))
```

# Comment afficher des domaines

```python
plot_implicit(And(x ** 2 + y ** 2 < 1, x > 0, y > 0))
```

# Comment afficher le graphe d'une fonction de deux variables

```python
plotting.plot3d((2 + cos(10 * x)) * (y ** 2), (x, -1, 1), (y, -1, 1))
```

# Comment afficher une courbe paramétrée 3d

```python
plotting.plot3d_parametric_line(cos(x), sin(x), x, (x, -10, 10))
```

# Comment afficher une surface paramétrée

```python
plotting.plot3d_parametric_surface(x + y, x - y, x * y, (x, -1, 1), (y, -1, 1))
```
