````python
f,ax = plt.subplots(1,1);
x = np.linspace(0.0, 2*np.pi, 201)
ligne, = ax.plot(x, np.sin(x+0.5), lw=2.0);
@interact(alpha=0.5)
def affichage(alpha):
    ligne.set_ydata(np.sin(x+alpha))
    return f
````

````python
fig ,ax = plt.subplots(1,1);
x = np.linspace(0.0, 2*np.pi, 201)
ligne, = ax.plot(x, np.sin(x+0.1), lw=2.0);
ligne.set_label('${0}\sin(x+{1})$'.format(1.0, 0.5))
ax.legend()
ax.set_ylim(-10.0, 10.0)
ax.set_xlim(0.0, 2*np.pi)
@interact(alpha=(0.1, 5.0, 0.2), ampl=(1.0, 10.0, 0.5))
def affichage(alpha, ampl):
    ligne.set_ydata(ampl*np.sin(x+alpha))
    ligne.set_label('${0}\sin(x+{1:.2})$'.format(ampl, alpha))
    ax.legend()
    return fig
````
